```python
data.head()
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# 1️⃣ **Codificación de variables categóricas (One-Hot Encoding)**
encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False)  # Ignorar valores desconocidos

# Seleccionar columnas categóricas
X_train_cat = data.select_dtypes(["category", "object"])

# Aplicar One-Hot Encoding
X_train_cat_1hot = encoder.fit_transform(X_train_cat)

# Obtener nombres de las categorías codificadas
categories = encoder.categories_
X_train_cat_1hot_df = pd.DataFrame(X_train_cat_1hot, columns=encoder.get_feature_names_out(X_train_cat.columns), index=data.index)

# 2️⃣ **Estandarización de variables numéricas (StandardScaler)**
scaler = StandardScaler()

# Seleccionar columnas numéricas
X_train_num = data.select_dtypes(["float64", "int64"])

# Aplicar estandarización
X_train_num_std = scaler.fit_transform(X_train_num)

# Convertir a DataFrame
X_train_num_std_df = pd.DataFrame(X_train_num_std, columns=X_train_num.columns, index=data.index)

# 3️⃣ **Unir las variables transformadas**
data_transformed = pd.concat([X_train_num_std_df, X_train_cat_1hot_df], axis=1)

# Verificar el resultado final
print(data_transformed.head())
```
